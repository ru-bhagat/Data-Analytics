## Regression Analysis


### 1. Simple Linear Regression

Educators want to know if previous academic performance is often a strong indicator of future academic success.<br> <br>
**a)** Create a scatter plot between the dependent variable-Previous_Exam_Score and independent variable-Academic_Performance_Score with the best-fit line passing through.            (1 mark)

```r
if (!requireNamespace("ggplot2", quietly = TRUE)){
    install.packages("ggplot2")
}
library(ggplot2)
```

```r
ggplot(data, aes(x = Previous_Exam_Scores, y = Academic_Performance_Score)) +
  geom_point(alpha = 0.5) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Previous Exam Scores vs Academic Performance",
       x = "Previous Exam Scores",
       y = "Academic Performance Score") +
  theme_minimal()
```

**Question : What do you infer from the graph?**

**Answer :**

There is a clear linear relationship between Previous Score and Performance Index, indicating a strong positive correlation. As the Previous Score increases, the Performance Index tends to rise correspondingly. While this suggests that a student's previous score is a reliable predictor of their performance index, the variability observed in the data indicates that it is not a flawless predictor.

**b)** Build a linear regression model to predict performance index based on previous scores  (1 mark).

```r
model <- lm(Academic_Performance_Score ~ Previous_Exam_Scores, data = data)
summary(model)
```

**c)** Check if this model follows OLS assumptions.              (3 marks)


```r
hist(model$residuals, main = "Histogram of Residuals", xlab = "Residuals", 
     col = "lightblue", border = "black", breaks = 20)

plot(model, which = 1)
qqnorm(model$residuals)
qqline(model$residuals)
```

**Question :**

Does the model follow OLS assumptions?          
How did you evaluate whether the linear regression model meets the assumptions required for Ordinary Least Squares (OLS) estimation?

**Answer :**

Yes, the model satisfies the assumptions of Ordinary Least Squares (OLS). The regression model is linear in its parameters, and the residuals follow a normal distribution, as demonstrated by the histogram and Q-Q plot. Additionally, the residuals vs. fitted values plot confirms homoscedasticity, as there is no funnel shape or any systematic pattern indicating changing variance.

**d)** SLR Validation:                          (1 mark) <br>
Evaluate the overall goodness-of-fit of the linear regression model.
Test whether the overall model is statistically significant using F-statistic.Mention the null and alternative hypotheses.

```r
model_summary <- summary(model)
print(model_summary)

f_statistic <- model_summary$fstatistic[1]
f_p_value <- pf(f_statistic, df1 = model_summary$fstatistic[2], df2 = model_summary$fstatistic[3], lower.tail = FALSE)

cat("F-statistic:", f_statistic, "\n")
cat("p-value for F-statistic:", f_p_value, "\n")
```

**Goodness-of-Fit:**

The R-squared value is 0.8376, meaning approximately 83.76% of the variance in Academic_Performance_Score is explained by Previous_Exam_Scores, which is considered a very good fit.

**Statistical Significance:**

The F-statistic is 51555.11 with a p-value of 0, indicating that the model is statistically significant and Previous_Exam_Scores significantly explains the variance in Academic_Performance_Score.

**Hypotheses for F-Test:**

Null Hypothesis (H0): The coefficient for Previous_Exam_Scores is equal to zero, implying it does not significantly affect Academic_Performance_Score.

Alternative Hypothesis (H1): The coefficient for Previous_Exam_Scores is not equal to zero, implying it significantly affects Academic_Performance_Score.

Given the F-statistic and p-value, we can conclude that the model is statistically significant.
